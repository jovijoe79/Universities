```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\University and vocational schools admission in Russia 2014-2023.csv")
df
```

```python
# FIRST WE TRY RO PRESERVE DATA INTEGRITY
df2 = df
df2
```

```python
# NEXT WE DROP DUPLICATES
df2.drop_duplicates()
```

```python
# NOW WE GO INTO DATA CLEANING PROPERLY
df2.info()
```

```python
df2
```

```python
df2 = df2.dropna()
df2
```

```python
df3 = df2
df3
```

```python
df2 = df2[(df2['Number of Applications'] != 0.0) & (df2['Number of Students'] != 0.0)].reset_index(drop= True)
df2
```

```python
# EXPLORATORY ANALYSIS
# 1) COUNT OF EDUCATION LEVEL
educ_level = df2[df2['Education level'] != ''].groupby('Education level')
educ_level = educ_level.size().reset_index(name= 'Count').sort_values(by='Count', ascending= False)
educ_level
```

```python
# COUNT OF DEGREE
Degree = df2[df2['Degree'] != ''].groupby('Degree')
Degree = Degree.size().reset_index(name= 'Count').sort_values(by='Count', ascending= False)
Degree
```

```python
# COUNT OF TUITION FEES
tuition = df2[df2['Tuition fees'] != ''].groupby('Tuition fees')
tuition = tuition.size().reset_index(name= 'Count').sort_values(by='Count', ascending= False)
tuition
```

```python
# Count of branches
branches = df2[df2['Branches of Science'] != ''].groupby('Branches of Science')
branches = branches.size().reset_index(name= 'Count').sort_values(by='Count', ascending= False)
branches
```

```python
# COUNT OF PROFESSION
professions = df2[df2['Group of Professions'] != ''].groupby('Group of Professions')
professions = professions.size().reset_index(name= 'Count').sort_values(by='Count', ascending= False)
professions
```

```python
#COUNT OF YEAR
years = df2[df2['Year'] != ''].groupby('Year')
years = years.size().reset_index(name= 'Count').sort_values(by='Count', ascending= False)
years
```

```python
# What profession got the most number of students
most_students = df2[(df2['Group of Professions'] != '') & (df2['Number of Students'] != '')]
most_students = most_students.groupby('Group of Professions')['Number of Students']
most_students = most_students.sum().reset_index(name= 'total').sort_values(by='total', ascending= False)
most_students
```

```python
# What profession got the most number of applications
most_applications = df2[(df2['Group of Professions'] != '') & (df2['Number of Applications'] != '')]
most_applications = most_applications.groupby('Group of Professions')['Number of Applications']
most_applications = most_applications.sum().reset_index(name= 'total').sort_values(by='total', ascending= False)
most_applications
```

```python
# In 2020 what profession had the most application 
most_applications = df2[(df2['Group of Professions'] != '') & (df2['Number of Applications'] != '') & (df2['Year'] == 2020.0)]
most_applications = most_applications.groupby('Group of Professions')['Number of Applications']
most_applications = most_applications.sum().reset_index(name= 'total').sort_values(by='total', ascending= False)
most_applications
```

```python
# In 2020 what profession had the most students
most_applications = df2[(df2['Group of Professions'] != '') & (df2['Number of Students'] != '') & (df2['Year'] == 2020.0)]
most_applications = most_applications.groupby('Group of Professions')['Number of Students']
most_applications = most_applications.sum().reset_index(name= 'total').sort_values(by='total', ascending= False)
most_applications
```
